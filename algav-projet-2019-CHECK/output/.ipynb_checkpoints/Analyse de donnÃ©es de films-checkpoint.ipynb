```python
# Faire un dictionnaire avec tous les acteurs (acteur => indice)
# + un dictionnaire inversé (indice => acteur)
ACTORS = dict()
ACTORS_INV = dict()
for lista in plusActors:
    for a in lista:
        # affecte une valeur à une clé si la clé n'est pas utilisée
        res = ACTORS.setdefault(a['name'], len(ACTORS))
        if res == len(ACTORS)-1:
            ACTORS_INV[len(ACTORS)-1] = a['name']
```

```python
#Dictionnaire tel que:
#key : genre de film (string)
#value : ensemble des films (id) appartenant au genre (tab de int)
MOVIESSORTEDBYGENRE = {}
NB_FILM = 0

for index, row in movies.iterrows():
    NB_FILM+=1
    listGenre = row['genres'].split("|")
    for i in listGenre:
        if i not in MOVIESSORTEDBYGENRE:
            MOVIESSORTEDBYGENRE[i] = []
        MOVIESSORTEDBYGENRE[i].append(row['movieId'])
        
# Liste des genres disponibles dans la base de donnée
ALLGENRE = MOVIESSORTEDBYGENRE.keys()
```

```python
# Filtre une liste de film à partir d'un genre
def filterListMovieByGenre(listMovieId, genre):
    res = []
    for i in listMovieId:
            if i in MOVIESSORTEDBYGENRE[genre]:
                res.append(i)
    return res

# Retourne le genre majoritaire d'un acteur
def mainGenreOfActor(actor):
    listMovies = MOVIESOFACTOR[actor]
    #dico tq: genre -> nombre de film du genre
    maxNb = 0
    maxGenre = "None"
    for genre in list(ALLGENRE):
        temp = len(filterListMovieByGenre(listMovies, genre))
        if temp > maxNb:
            maxNb = temp
            maxGenre = genre
            
    return maxGenre


def genresOfMovie(idFilm):
    return movies[movies.movieId == idFilm].genres.iloc[0].split("|")

# Retourne la liste des films dans lesquels l'acteur passé en paramètre joue
# Un genre peut être spécifié pour filtrer les films

def filmsOfActor(acteur, genre='all'):
    #Recuperation des index des films où joue acteur dans plusActors
    listMovie = MOVIESOFACTOR[acteur]
        
    #Si pas de genre spécifié, on retourne l'ensemble des films
    if genre == 'all':
        return listMovie
    else:
        return filterListMovieByGenre(listMovie, genre)

def filmsOfProd(prod, genre='all'):
    #Recuperation des index des films où joue acteur dans plusActors
    listMovie = MOVIESOFPROD[prod]
        
    #Si pas de genre spécifié, on retourne l'ensemble des films
    if genre == 'all':
        return listMovie
    else:
        return filterListMovieByGenre(listMovie, genre)
    
param_Actor = "Tom Hanks"
param_Genre = "Comedy"

print(param_Actor, "\n\t", len(filmsOfActor(param_Actor)), "film(s) au total,\n\t",
    len(filmsOfActor(param_Actor,param_Genre)), "dans le genre", param_Genre)

```

```python
#Retourne une liste de movieId qui ont le genre voulu
def getIdFromGenre(liste, genre):
    rList=[]
    for index, row in liste.iterrows():
        if (genre in row['genres'].split('|')):
            rList.append(row['movieId'])
        df=pd.DataFrame({'movieId':rList})
    return df

#Retourne la note moyenne d'un film
def getRating(liste, id):
    return liste[liste.movieId==id]['rating'].mean()

#Retourne les films vu selon l'userId
def getVue(liste, userId):
    return liste[liste.userId==userId]

#Retourne le titre du film selon l'id
def getTitleById(mId):
    return movies[movies.movieId == mId]['title'].iloc[0]

#Sépare un LabeledSet en deux en fonction du ratio passé en paramètre
def split(labeledSet,ratio):
    """ Separe en deux le labeledSet par un ratio donnés en paramètre.
        Hypothese: 0 < ratio < 1
    """
    #tableau contenant les index 
    index = []
    for i in range(labeledSet.size()):
        index.append(i)
    #mélange des index
    shuffle(index)
    
    dim = labeledSet.getInputDimension()
    #creation des deux labeledSet
    lS1 =ls.LabeledSet(dim)  
    lS2 =ls.LabeledSet(dim)  
    
    #ajout
    for j in range (int(labeledSet.size()*ratio)):
        i = index[j]
        lS1.addExample(labeledSet.getX(i),labeledSet.getY(i))
        
    for j in range (int(labeledSet.size()*ratio)+1,labeledSet.size()):
        i = index[j]
        lS2.addExample(labeledSet.getX(i),labeledSet.getY(i))
        
    return lS1,lS2

```

# Classifiers

### Prédiction de note

On crée un classifieur GradiantBatch adapté:

```python
from random import shuffle

class ClassifierGradiantBatchFilm(cl.Classifier):
    def __init__(self,input_dimension,learning_rate):
        """ Argument:
                - intput_dimension (int) : dimension d'entrée des exemples
                - learning_rate :
            Hypothèse : input_dimension > 0
        """
        self.alpha = learning_rate
        self.input_dimension = input_dimension
        
        v = np.random.rand(input_dimension)
        self.w = np.linalg.norm(v)
        
    def predict(self,x):
        """ rend la prédiction de la note
        """
        return np.dot(x, self.w)
    
    def train(self,labeledSet):
        """ Permet d'entrainer le modele sur l'ensemble donné
        """
        #tableau contenant les index 
        index = []
        for i in range(labeledSet.size()):
            index.append(i)
        #mélange des index
        shuffle(index)
        
        g = 0
        for i in range(len(index)):
            xi = labeledSet.getX(index[i])
            yi = labeledSet.getY(index[i])
            g = g +(yi - np.dot(xi,self.w))*xi
        self.w = self.w + self.alpha * g
        
        
    def accuracy(self, dataset):
        """ Permet de calculer la qualité du système
            Retourne l'ecart moyen
        """
        cpt=0
        for i in range(0,dataset.size()):
            cpt+=abs(self.predict(dataset.getX(i))-dataset.getY(i)[0])
        return float(cpt)/dataset.size()
```

Fonction pour factoriser le code

```python
#Train le classifieur sur les données passées en paramètre et affiche l'évolution de l'accuracy
def trainAndShowEvolution(cla, ls, ratio, iterationTrain, xlabel="k", ylabel="accuracy",
            title="Evolution de l'accuracy en fonction du nombre d'itération"):
    
    #séparation des données en deux ensembles: train et test
    dataTrain, dataTest = split(ls,ratio)
    
    #tableaux pour stocker l'accuracy en fonction du nombre d'itération
    trainRes = []
    testRes = []
    
    #premier train
    #cla.train(dataTrain)
    trainRes.append(cla.accuracy(dataTrain))
    testRes.append(cla.accuracy(dataTest))
    
    #iterationTrain trains réalisés
    for i in range(iterationTrain):
        cla.train(dataTrain)
        trainRes.append(cla.accuracy(dataTrain))
        testRes.append(cla.accuracy(dataTest))
    
    #affiche des courbes montrant l'évolution de l'accuracy en fontion du nombre d'itération
    plt.ylabel(ylabel)
    plt.xlabel(xlabel)
    plt.title(title)

    legendtrain, = plt.plot(trainRes,label="train")
    legendtest, = plt.plot(testRes,label="test")
    plt.legend(handles=[legendtrain, legendtest])
```

# Arbre de décision

```python
cpt=1
enumGenre = {}
for genre in list(ALLGENRE):
    enumGenre[genre] = cpt
    cpt+=1

def getGenreByEnum(value):
    for key, v in enumGenre.items():
        if value == v:
            return key
    return "null"
```

```python
def shannon(P):
    somme = 0;
    
    for i in P:
        if (i != 0 and not (i==1.0 and len(P)==1)):
            somme = somme + i * math.log(i,len(P));
    
    return -somme;

def entropie(labeledSet) :
    YwithOccurence = {}

    for i in range(labeledSet.size()):    
        current_y = labeledSet.getY(i)[0]
        if current_y not in YwithOccurence:
            YwithOccurence[current_y] = 0
        YwithOccurence[current_y] += 1
    
    tab = []
   
    for valeur in YwithOccurence.values():
        tab.append(valeur/labeledSet.size())
    return shannon(tab)
def divise(LSet, att, seuil):
    #doit renvoyer deux labeledSet separes par le seuil sur la colonne att
    inf = ls.LabeledSet(LSet.getInputDimension()) #ensemble E1
    sup = ls.LabeledSet(LSet.getInputDimension()) #ensemble E2
    #print LSet.size()
    for i in range(LSet.size()):
#         print LSet.getY(i)[0]
        if(LSet.getX(i)[att] > seuil):
            sup.addExample(LSet.getX(i), LSet.getY(i))
        else:
            inf.addExample(LSet.getX(i), LSet.getY(i))
    return inf, sup

class Arbre:
    def __init__(self):
        self.attribut = None   # numéro de l'attribut
        self.seuil = None
        self.inferieur = None # ArbreBinaire Gauche (valeurs <= au seuil)
        self.superieur = None # ArbreBinaire Gauche (valeurs > au seuil)
        self.classe = None # Classe si c'est une feuille: -1 ou +1
        
    def est_feuille(self):
        """ rend True si l'arbre est une feuille """
        return self.seuil == None
    
    def ajoute_fils(self,ABinf,ABsup,att,seuil):
        """ ABinf, ABsup: 2 arbres binaires
            att: numéro d'attribut
            seuil: valeur de seuil
        """
        self.attribut = att
        self.seuil = seuil
        self.inferieur = ABinf
        self.superieur = ABsup
    
    def ajoute_feuille(self,classe):
        """ classe: -1 ou + 1
        """
        self.classe = classe
        
    def classifie(self,exemple):
        """ exemple : numpy.array
            rend la classe de l'exemple: +1 ou -1
        """
        if self.est_feuille():
            return self.classe
        if exemple[self.attribut] <= self.seuil:
            return self.inferieur.classifie(exemple)
        return self.superieur.classifie(exemple)
    
    def to_graph(self, g, prefixe='A'):
        """ construit une représentation de l'arbre pour pouvoir
            l'afficher
        """
        if self.est_feuille():
            g.node(prefixe,str(self.classe),shape='box')
        else:
            g.node(prefixe, str(self.attribut))
            self.inferieur.to_graph(g,prefixe+"g")
            self.superieur.to_graph(g,prefixe+"d")
            g.edge(prefixe,prefixe+"g", '<='+ str(self.seuil))
            g.edge(prefixe,prefixe+"d", '>'+ str(self.seuil))
        
        return g

def discretise(LSet, col, classe):
    """ LabelledSet * int -> tuple[float, float]
        Hypothèse: LSet.size() >= 2
        col est le numéro de colonne sur X à discrétiser
        rend la valeur de coupure qui minimise l'entropie ainsi que son entropie.
    """
    # initialisation:
    min_entropie = 1.1  # on met à une valeur max car on veut minimiser
    min_seuil = 0.0     
    # trie des valeurs:
    ind= np.argsort(LSet.x,axis=0)
    
    # calcul des distributions des classes pour E1 et E2:
    inf_plus  = 0               # nombre de +1 dans E1
    inf_moins = 0               # nombre de -1 dans E1
    sup_plus  = 0               # nombre de +1 dans E2
    sup_moins = 0               # nombre de -1 dans E2       
    # remarque: au départ on considère que E1 est vide et donc E2 correspond à E. 
    # Ainsi inf_plus et inf_moins valent 0. Il reste à calculer sup_plus et sup_moins 
    # dans E.
    for j in range(0,LSet.size()):
        if (LSet.getY(j) == classe):
            sup_moins += 1
        else:
            sup_plus += 1
    nb_total = (sup_plus + sup_moins) # nombre d'exemples total dans E
    
    # parcours pour trouver le meilleur seuil:
    for i in range(len(LSet.x)-1):
        v_ind_i = ind[i]   # vecteur d'indices
        courant = LSet.getX(v_ind_i[col])[col]
        lookahead = LSet.getX(ind[i+1][col])[col]
        val_seuil = (courant + lookahead) / 2.0;
        # M-A-J de la distrib. des classes:
        # pour réduire les traitements: on retire un exemple de E2 et on le place
        # dans E1, c'est ainsi que l'on déplace donc le seuil de coupure.
        if LSet.getY(ind[i][col])[0] == classe:
            inf_moins += 1
            sup_moins -= 1
        else:
            inf_plus += 1
            sup_plus -= 1
        # calcul de la distribution des classes de chaque côté du seuil:
        nb_inf = (inf_moins + inf_plus)*1.0     # rem: on en fait un float pour éviter
        nb_sup = (sup_moins + sup_plus)*1.0     # que ce soit une division entière.
        # calcul de l'entropie de la coupure
        val_entropie_inf = shannon([inf_moins / nb_inf, inf_plus  / nb_inf])
        val_entropie_sup = shannon([sup_moins / nb_sup, sup_plus  / nb_sup])
        val_entropie = (nb_inf / nb_total) * val_entropie_inf \
                       + (nb_sup / nb_total) * val_entropie_sup
        # si cette coupure minimise l'entropie, on mémorise ce seuil et son entropie:
        if (min_entropie > val_entropie):
            min_entropie = val_entropie
            min_seuil = val_seuil

    return (min_seuil, min_entropie)

def LSetequal(LSet1,LSet2):
    if LSet1.size() != LSet2.size():
        return False
    
    for i in range(LSet1.size()):
        if len(LSet1.getX(i)) != len(LSet2.getX(i)):
            print("2")
            return False
        for j in range(len(LSet1.getX(i))):
            if LSet1.getX(i)[j] != LSet2.getX(i)[j]:
                print("3")
                print(LSet1.getX(i)[j],LSet2.getX(i)[j])
                return False
        
    return True
            
    
def construit_AD(LSet,epsilon):
    arbre = Arbre()
   
    if(LSet.size()>0):
        entrop = entropie(LSet)
        if (entrop <= epsilon):
            arbre.ajoute_feuille(classe_majoritaire(LSet))
        else:
            imin = 0
            entropmin = 1.1
            seuilmin =0
            cmaj=classe_majoritaire(LSet) # à modifier
           
            for i in range(LSet.getInputDimension()):
                (seuil, e) = discretise(LSet,i, cmaj)
                if e < entropmin:
                    entropmin = e
                    imin = i
                    seuilmin = seuil
            
            l1,l2 = divise(LSet,imin,seuilmin)
            
            finished = False
            if LSetequal(LSet, l1):
                arbre.ajoute_feuille(classe_majoritaire(l1))
                finished = True
            if LSetequal(LSet, l2):
                arbre.ajoute_feuille(classe_majoritaire(l2))
                finished = True
            
            if finished:
                return arbre
            
            arbre.ajoute_fils(construit_AD(l1,epsilon), construit_AD(l2,epsilon),imin,seuilmin)
        
    return arbre

"""Renvoie la classe la plus présente dans le labeledset donné"""
def classe_majoritaire(labeledSet):
    YwithOccurence = {}

    for i in range(labeledSet.size()):  
        current_y = labeledSet.getY(i)[0]
        if current_y not in YwithOccurence:
            YwithOccurence[current_y] = 0
        YwithOccurence[current_y] += 1
    
    cmaj=-1
    for key, value in YwithOccurence.items():
        if ((cmaj== -1) or (value>=cmaj)):
            cmaj=key
    return cmaj

class ArbreDecision(cl.Classifier):
    # Constructeur
    def __init__(self,epsilon):
        # valeur seuil d'entropie pour arrêter la construction
        self.epsilon= epsilon 
        self.racine = None
    
    # Permet de calculer la prediction sur x => renvoie un score
    def predict(self,x):
        # classification de l'exemple x avec l'arbre de décision
        # on rend 0 (classe -1) ou 1 (classe 1)
        return self.racine.classifie(x)
        
    # Permet d'entrainer le modele sur un ensemble de données
    def train(self,set):
        # construction de l'arbre de décision 
        self.set=set
        self.racine = construit_AD(set,self.epsilon)

    # Permet d'afficher l'arbre
    def plot(self):
        gtree = gv.Digraph(format='png')
        return self.racine.to_graph(gtree)
    
    def accuracy(self,label):
        cpt=0
        for i in range(label.size()):
            if(self.predict(label.getX(i))==label.getY(i)):
                cpt+=1
        return cpt/label.size()
```

## Évaluation du clustering

```python
def dist_intracluster(df):
    d_max = 0
    for i in range(df.shape[0]):
        for j in range(df.shape[0]):
            d = km.dist_vect(df.iloc[i], df.iloc[j])
        if d > d_max:
            d_max = d
    return d_max

def global_intraclusters(df, affectation):
    dist_intra = []
    for i in affectation.values():
        dist_intra.append(dist_intracluster(df.iloc[i]))
    return max(dist_intra)

def sep_clusters(df_centres):
    d_min = 1
    size = df_centres.shape[0]
    for i in range(size):
        for j in range(size):
            if i != j:
                d = km.dist_vect(df_centres.iloc[i], df_centres.iloc[j])
                if d < d_min:
                    d_min = d
    return d_min

def evaluation(index, df, centres, affectation):
    if index == "Dunn":
        return global_intraclusters(df, affectation)/sep_clusters(centres)
    elif index == "XB":
        return km.inertie_globale(df, affectation)/sep_clusters(centres)
    else:
        print ("Mauvaise valeur pour index")
        
```

```python
def dist_manhattan_vect(x,y):
    ret=0
    if len(x) < len(y):
        for i in range(len(x),len(y)):
            x = np.append(x,0)
    else:
        for i in range(len(y),len(x)):
            y = np.append(y,0)
            
    for i in range(len(x)):
        ret+=math.fabs(x[i]-y[i])
    return ret

def dist_euclidienne_vect(x,y):
    ret=0
    if len(x) < len(y):
        for i in range(len(x),len(y)):
            x = np.append(x,0)
    else:
        for i in range(len(y),len(x)):
            y = np.append(y,0)

    for i in range(len(x)):
        ret+=(x[i]-y[i])**2
    return math.sqrt(ret)

def dist_vect(option,x,y):
    if (option == "euclidienne"):
        return dist_euclidienne_vect(x,y)
    return dist_manhattan_vect(x,y)
def normalisation(dataFrame):
    
    data=dataFrame.as_matrix()
    for i in range(data.shape[1]):
        minCur=data[:,i].min()
        maxCur=data[:,i].max()
        
        data[:,i]=(data[:,i]-minCur)/(maxCur-minCur)
         
    return data
def centroide(matrix_data):
    
    res=np.array([])
    
    if len(np.shape(matrix_data)) > 1:
        for i in range(len(matrix_data[0])):
            #print("le moyenne pour\n",matrix_data)
            res=np.append(res,np.mean(matrix_data[:,i]))
        return res
    return matrix_data

def dist_groupes(option,g1,g2):
    
    c1=centroide(np.array(g1))
    c2=centroide(np.array(g2))
    
    return dist_vect(option,c1,c2)

def initialise(M):
    res = {}
    cpt = 0
    for i in M:
        res[cpt] = i
        cpt+=1
    return res

def dist_max_groupes(option,g1,g2):
    dmax=0
    dcur=0
    for e1 in g1:
        for e2 in g2:
            dcur=dist_vect(option,e1,e2)
            if dmax<dcur:
                dmax=dcur
    return dmax

def fusionneComplete(option, C0):
    
    dmin=100000
    dCur=2
    ind1=-1
    ind2=-1
    for i in C0.keys():
        for j in C0.keys():
            if j!=i: 
                dCur=dist_max_groupes(option,C0[i],C0[j])
                if(dCur < dmin):
                    dmin=dCur
                    ind1=i
                    ind2=j
    C1=dict()
    der=-2
    
    for i in C0.keys():
        if i != ind1 or i != ind2:
            C1[i]=C0[i]
            der=i
            
        
    
    der+=1
    #tmp=[C0[ind1][0]]
    
    tmp=np.vstack([C0[ind1][0],C0[ind2][0]])
    C1[der]=np.array(tmp)

    del C1[ind1]
    del C1[ind2]
    print("fusion de ",ind1," et ",ind2,"pour une distance de ",dmin)
    return C1,ind1,ind2,dmin

def fusionne(chaine,data):
    distMin = -1
    indexTuple = [-1,-1]
    indexMax = 0
    
    for index1,vec1 in data.items():
        for index2,vec2 in data.items():
            if (index1 == index2):
                continue
            
            dist = dist_groupes(chaine, [vec1], [vec2])
            if (dist < distMin or distMin == -1):
                distMin = dist;
                indexTuple[0] = index1
                indexTuple[1] = index2
        if (indexMax < index1):
            indexMax = index1
    
    vec1 = data[indexTuple[0]]
    vec2 = data[indexTuple[1]]
    
    newVec = []
    for i in vec1:
        newVec.append(i)
    for i in vec2:
        newVec.append(i)
    
    data[indexMax+1] = np.array(newVec)

   
    del data[indexTuple[0]]
    del data[indexTuple[1]]
    
    
    return data,indexTuple[0],indexTuple[1],distMin
def clustering_hierarchique(data, option):
    #initialisation 
    courant = initialise(data)       # clustering courant, au départ:s données data_2D normalisées
    mat_fusion = []  # initialisation
    i=0
    while len(courant) >=2:              # tant qu'il y a 2 groupes à fusionner
        #print("fusionne",i)
        new,k1,k2,dist_min = fusionne(option,courant)
        if(len(mat_fusion)==0):
            mat_fusion = [k1,k2,dist_min,2]
        else:
            mat_fusion = np.vstack( [mat_fusion,[k1,k2,dist_min,2] ])
        courant = new
        i+=1
        
        
    # Paramètre de la fenêtre d'affichagemat_fusion
    plt.figure(figsize=(30, 15)) # taille : largeur x hauteur
    plt.title('Dendrogramme', fontsize=25)    
    plt.xlabel('Exemple', fontsize=25)
    plt.ylabel('Distance', fontsize=25)

    # Construction du dendrogramme à partir de la matrice M_Fusion:
    scipy.cluster.hierarchy.dendrogram(
        mat_fusion,
        leaf_font_size=18.,  # taille des caractères de l'axe des X
    )

    # Affichage du résultat obtenu:
    plt.show()
    return mat_fusion

def clustering_hierarchiqueCompleteLinkage(data, option):
    #initialisation 
    courant = initialise(data)       # clustering courant, au départ:s données data_2D normalisées
    mat_fusion = []  # initialisation
    i=0
    while len(courant) >=2:              # tant qu'il y a 2 groupes à fusionner
        #print("fusionne",i)
        new,k1,k2,dist_min = fusionneComplete(option,courant)
        if(len(mat_fusion)==0):
            mat_fusion = [k1,k2,dist_min,2]
        else:
            mat_fusion = np.vstack( [mat_fusion,[k1,k2,dist_min,2] ])
        courant = new
        i+=1
        
        
    # Paramètre de la fenêtre d'affichagemat_fusion
    plt.figure(figsize=(30, 15)) # taille : largeur x hauteur
    plt.title('Dendrogramme', fontsize=25)    
    plt.xlabel('Exemple', fontsize=25)
    plt.ylabel('Distance', fontsize=25)

    # Construction du dendrogramme à partir de la matrice M_Fusion:
    scipy.cluster.hierarchy.dendrogram(
        mat_fusion,
        leaf_font_size=18.,  # taille des caractères de l'axe des X
    )

    # Affichage du résultat obtenu:
    plt.show()
    return mat_fusion
```
